```python
# Definiamo il percorso del file
file_path = r"D:\python_project\imitation_data_complete.xlsx"

# Carichiamo l'Excel specificando il foglio corretto
# Usiamo 'data_pcr' come avevamo stabilito per l'analisi
df = pd.read_excel(file_path, sheet_name='data_pcr')

# Visualizziamo le prime 5 righe per controllare che sia tutto ok
df
```

```python
# Rimuoviamo righe dove mancano i dati fondamentali (DCt o Fold_Change)
df_clean = df.dropna(subset=['DCt', 'Fold_Change']).copy()

# Definiamo l'ordine logico delle condizioni e delle aree
# Questo servirà per avere i grafici ordinati correttamente
aree_ordine = ["Wulst", "MSt", "IMM", "Nidopallium", "Arcopallium"]
cond_ordine = ["A", "O", "C"]
geni_ordine = ["cFOS", "EGR1", "ARC"]

# Trasformiamo le colonne in "Categorical" (simile ai factor di R)
df_clean['Condition'] = pd.Categorical(df_clean['Condition'], categories=cond_ordine, ordered=True)
df_clean['Area'] = pd.Categorical(df_clean['Area'], categories=aree_ordine, ordered=True)
df_clean['Gene'] = pd.Categorical(df_clean['Gene'], categories=geni_ordine, ordered=True)

df_clean
```

```python
import pandas as pd

# Caricamento dei dati
filename = "imitation_data_complete.xlsx"
data_behav = pd.read_excel(filename, sheet_name="data_behav")
data_pcr = pd.read_excel(filename, sheet_name="data_pcr")

# Summary tables (Conteggio SubID unici per condizione)
behav_summary = data_behav.groupby('Condition')['SubID'].nunique().reset_index(name='n_chicks')
pcr_summary = data_pcr.groupby('Condition')['SubID'].nunique().reset_index(name='n_chicks')

print("\n--- Summary for Behavioral Data ---")
print(behav_summary)

print("\n--- Summary for PCR Data ---")
print(pcr_summary)
```

```python
# Calcoliamo Media, Deviazione Standard e Errore Standard (SEM) per ogni gruppo
stats_summary = df_clean.groupby(['Area', 'Gene', 'Condition'], observed=True).agg({
    'DCt': ['mean', 'std', 'count'],
    'Fold_Change': ['mean', 'std', 'count']
}).reset_index()

# Calcolo del SEM (Standard Error of the Mean)
stats_summary[('DCt', 'sem')] = stats_summary[('DCt', 'std')] / np.sqrt(stats_summary[('DCt', 'count')])
stats_summary[('Fold_Change', 'sem')] = stats_summary[('Fold_Change', 'std')] / np.sqrt(stats_summary[('Fold_Change', 'count')])
# Rende la tabella più leggibile
stats_summary.columns = ['_'.join(col).strip() if col[1] else col[0] for col in stats_summary.columns]
stats_summary
```

```python
from scipy import stats
import pandas as pd

# 1. Filtriamo per includere solo i tre gruppi principali (A, O, C)
df_filtered = df_clean[df_clean['Condition'].isin(['A', 'O', 'C'])].copy()

normality_and_variance = []

# Raggruppiamo per Area e Gene
for (area, gene), group in df_filtered.groupby(['Area', 'Gene'], observed=True):
    
    # --- A. TEST DI NORMALITÀ GLOBALE (Per Area e Gene, tutte le condizioni insieme) ---
    _, p_shapiro_global = stats.shapiro(group['DCt'])
    
    # --- B. TEST DI NORMALITÀ PER OGNI CONDIZIONE (A, O, C separati) ---
    cond_p_values = {}
    for cond in ['A', 'O', 'C']:
        subset = group[group['Condition'] == cond]['DCt']
        if len(subset) > 2:
            _, p_shapiro = stats.shapiro(subset)
            cond_p_values[f'Shapiro_p_{cond}'] = round(p_shapiro, 4)
        else:
            cond_p_values[f'Shapiro_p_{cond}'] = None

    # --- C. TEST OMOGENEITÀ VARIANZA (Levene) TRA I 3 GRUPPI ---
    data_groups = [group[group['Condition'] == c]['DCt'].values for c in ['A', 'O', 'C']]
    data_groups = [g for g in data_groups if len(g) > 0]
    
    if len(data_groups) > 1:
        _, p_levene = stats.levene(*data_groups)
    else:
        p_levene = None

    # --- RACCOLTA RISULTATI ---
    normality_and_variance.append({
        'Area': area,
        'Gene': gene,
        'p_Shapiro_Global': round(p_shapiro_global, 4), # <--- Nuova Colonna
        'p_Shapiro_A': cond_p_values['Shapiro_p_A'],
        'p_Shapiro_O': cond_p_values['Shapiro_p_O'],
        'p_Shapiro_C': cond_p_values['Shapiro_p_C'],
        'p_Levene': round(p_levene, 4) if p_levene is not None else None,
        'Assunzioni_OK?': '✅' if all(p > 0.05 for p in [cond_p_values['Shapiro_p_A'], 
                                                        cond_p_values['Shapiro_p_O'], 
                                                        cond_p_values['Shapiro_p_C'], 
                                                        p_levene] if p is not None) else '❌'
    })

# Creazione DataFrame finale
df_assunzioni_full = pd.DataFrame(normality_and_variance)

print("--- VERIFICA ASSUNZIONI ANOVA (Con Normalità Globale per Area/Gene) ---")
display(df_assunzioni_full)
```

```python
import scikit_posthocs as sp # Se dà errore, scrivi !pip install scikit-posthocs in una cella

kruskal_results = []

for (area, gene), group in df_clean.groupby(['Area', 'Gene'], observed=True):
    # Prepariamo i dati per i tre gruppi
    data_groups = [group[group['Condition'] == c]['DCt'] for c in cond_ordine]
    
    # Kruskal-Wallis test
    h_stat, p_kruskal = stats.kruskal(*data_groups)
    
    # Se p_kruskal < 0.05, facciamo il Post-hoc di Dunn
    posthoc_res = "N/A"
    if p_kruskal < 0.05:
        # Dunn test con correzione di Bonferroni
        dunn_df = sp.posthoc_dunn(group, val_col='DCt', group_col='Condition', p_adjust='bonferroni')
        # Estraiamo i p-value che ci interessano
        posthoc_res = f"A-C: {dunn_df.loc['A','C']:.3f}, O-C: {dunn_df.loc['O','C']:.3f}"

    kruskal_results.append({
        'Area': area, 'Gene': gene, 'Kruskal_p': round(p_kruskal, 4),
        'PostHoc_Dunn': posthoc_res
    })

df_kruskal = pd.DataFrame(kruskal_results)
df_kruskal
```

```python
ttest_results = []

for (area, gene), group in df_clean.groupby(['Area', 'Gene'], observed=True):
    # Gruppi
    a = group[group['Condition'] == 'A']['DCt']
    o = group[group['Condition'] == 'O']['DCt']
    c = group[group['Condition'] == 'C']['DCt']
    
    # Eseguiamo il T-test di Welch (non assume varianze uguali, è più sicuro)
    res_ac = stats.ttest_ind(a, c, equal_var=False)
    res_oc = stats.ttest_ind(o, c, equal_var=False)
    
    ttest_results.append({
        'Area': area, 'Gene': gene,
        'A_vs_C_p': round(res_ac.pvalue, 4),
        'O_vs_C_p': round(res_oc.pvalue, 4)
    })

df_ttest = pd.DataFrame(ttest_results)

# Evidenziamo anche le tendenze (p < 0.1)
def highlight_trends(val):
    if val < 0.05: return 'background-color: #ffcccc; font-weight: bold' # Rosso: Sig.
    if val < 0.10: return 'background-color: #ffffcc' # Giallo: Tendenza
    return ''

df_ttest.style.map(highlight_trends, subset=['A_vs_C_p', 'O_vs_C_p'])
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from IPython.display import display

# 1. PREPARAZIONE DATI (Solo A, O, C)
# Usiamo df_clean o il tuo dataset principale filtrato
df_lmm = df_clean[df_clean['Condition'].isin(['A', 'O', 'C'])].copy()

# 2. NORMALIZZAZIONE (Z-score per Gene)
# Questo rende i 3 geni confrontabili (media 0, deviazione standard 1)
df_lmm['DCt_norm'] = df_lmm.groupby('Gene')['DCt'].transform(lambda x: (x - x.mean()) / x.std())

print(f"\033[1m--- ANALISI MODELLI MISTI (LMM) - CONDIZIONI A, O, C ---\033[0m")
print("Fattore Between: Condition | Fattore Within: Gene | Effetto Random: SubID\n")

# Lista delle aree nell'ordine desiderato
aree_lista = df_lmm['Area'].unique()

for area in aree_lista:
    print(f"\n" + "="*50)
    print(f">>> AREA CEREBRALE: {area}")
    print("="*50)
    
    df_area = df_lmm[df_lmm['Area'] == area].copy()
    
    # Se un'area non ha abbastanza dati, la saltiamo
    if df_area.empty:
        continue

    # 3. DEFINIZIONE DEL MODELLO LINEARE A EFFETTI MISTI (LMM)
    # Formula: DCt_norm ~ Condition * Gene
    # groups="SubID" indica che le misure dei geni appartengono allo stesso individuo
    # Usiamo il Controllo (C) come riferimento (intercept)
    try:
        model = smf.mixedlm(
            "DCt_norm ~ C(Condition, Treatment('C')) * C(Gene)", 
            df_area, 
            groups=df_area["SubID"]
        )
        
        result = model.fit()
        
        # Estrazione della tabella dei risultati
        summary_df = result.summary().tables[1]
        summary_df['P>|z|'] = pd.to_numeric(summary_df['P>|z|'])
        
        # Formattazione: Rosso se p < 0.05
        def highlight_sig(val):
            return 'color: red; font-weight: bold' if val < 0.05 else 'color: black'

        display(summary_df.style.map(highlight_sig, subset=['P>|z|']))
        
        # 4. INTERPRETAZIONE RAPIDA
        print("\nInterpretazione Effetti Significativi (vs Controllo C):")
        found_sig = False
        
        for idx in summary_df.index:
            p_val = summary_df.loc[idx, "P>|z|"]
            if p_val < 0.05 and idx != "Intercept" and "Var" not in idx:
                coef = summary_df.loc[idx, "Coef."]
                print(f" • \033[91m{idx}\033[0m: p = {p_val:.4f} (Coef: {coef:.2f})")
                found_sig = True
        
        if not found_sig:
            print(" • Nessun effetto significativo trovato in quest'area.")

    except Exception as e:
        print(f"Errore nell'analisi dell'area {area}: {e}")
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
from IPython.display import display

# 1. Filtriamo per le condizioni principali (A, O, C)
df_lmm = df_clean[df_clean['Condition'].isin(['A', 'O', 'C'])].copy()

# 2. Lista dei geni presenti
geni_lista = df_lmm['Gene'].unique()

print(f"\033[1m--- ANOVA TABLE (WALD TEST) PER OGNI GENE ---\033[0m")
print("Modello: DCt ~ Condition * Area | Random: SubID\n")

for gene in geni_lista:
    print(f"\n" + "="*60)
    print(f"ANALISI PER IL GENE: {gene}")
    print("="*60)
    
    df_gene = df_lmm[df_lmm['Gene'] == gene].copy()
    
    if df_gene.empty:
        continue

    try:
        # Costruiamo il Modello a Effetti Misti per il singolo gene
        # L'interazione '*' tra Condition e Area è il cuore del test
        model = smf.mixedlm("DCt ~ C(Condition, Treatment('C')) * Area", 
                            df_gene, groups=df_gene["SubID"])
        result = model.fit()
        
        # Esecuzione del Wald Test sui termini del modello
        wald_test = result.wald_test_terms(scalar=False)
        anova_table = wald_test.table
        
        # Rinominia degli indici per chiarezza nella tesi
        new_indices = {
            "C(Condition, Treatment('C'))": "CONDITION (Effetto del Trattamento)",
            "Area": "AREA (Differenze tra Regioni)",
            "C(Condition, Treatment('C')):Area": "CONDITION x AREA (Interazione)"
        }
        anova_table.index = [new_indices.get(x, x) for x in anova_table.index]
        
        # Rimozione Intercetta per focus scientifico
        if "Intercept" in anova_table.index:
            anova_table = anova_table.drop("Intercept")

        # Formattazione: Rosso < 0.05
        def color_significant(val):
            return 'color: red; font-weight: bold' if val < 0.05 else 'color: black'

        # Visualizzazione Tabella
        styled_table = anova_table.style.map(color_significant, subset=['pvalue'])
        display(styled_table)

        # --- INTERPRETAZIONE RAPIDA ---
        p_cond = anova_table.loc["CONDITION (Effetto del Trattamento)", "pvalue"]
        p_area = anova_table.loc["AREA (Differenze tra Regioni)", "pvalue"]
        p_inter = anova_table.loc["CONDITION x AREA (Interazione)", "pvalue"]

        if p_inter < 0.05:
            print(f"\033[91m[!] INTERAZIONE SIGNIFICATIVA:\033[0m L'effetto di A e O su {gene} varia drasticamente tra le diverse aree.")
        elif p_cond < 0.05:
            print(f"\033[94m[*] EFFETTO CONDIZIONE:\033[0m Il trattamento ha un impatto generale su {gene} in tutto il cervello.")
        
        if p_area < 0.05:
            print(f" • Nota: Esistono differenze basali significative di {gene} tra le diverse aree.")

    except Exception as e:
        print(f"Errore nell'elaborazione del gene {gene}: {e}")
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
from IPython.display import display

# 1. Filtriamo per le condizioni principali (A, O, C)
df_lmm = df_clean[df_clean['Condition'].isin(['A', 'O', 'C'])].copy()

# 2. Normalizzazione DCt (Z-score per Gene)
df_lmm['DCt_norm'] = df_lmm.groupby('Gene')['DCt'].transform(lambda x: (x - x.mean()) / x.std())

# Definiamo l'ordine delle aree (usa quelle presenti nel tuo df)
aree_ordine = df_lmm['Area'].unique()

print(f"\033[1m--- TABELLA ANOVA (TYPE III WALD TEST) - CONDIZIONI UNITE (A, O, C) ---\033[0m")
print("Modello Misto: Condition (Between) + Gene (Within) | Random: SubID\n")

for area in aree_ordine:
    df_area = df_lmm[df_lmm['Area'] == area].copy()
    
    if df_area.empty:
        continue

    try:
        # Costruiamo il Modello a Effetti Misti (LMM)
        # L'interazione '*' permette di vedere se i geni rispondono diversamente alle condizioni
        model = smf.mixedlm("DCt_norm ~ C(Condition, Treatment('C')) * C(Gene)", 
                            df_area, groups=df_area["SubID"])
        result = model.fit()
        
        # Esecuzione del Wald Test per ottenere la tabella simile all'ANOVA
        # Questo test aggrega i coefficienti per darci un p-value unico per ogni fattore
        wald_test = result.wald_test_terms(scalar=False)
        anova_table = wald_test.table
        
        # Pulizia e rinomina degli indici per la tesi
        new_indices = {
            "C(Condition, Treatment('C'))": "CONDITION (Between: A, O, C)",
            "C(Gene)": "GENE (Within: ARC, EGR1, cFOS)",
            "C(Condition, Treatment('C')):C(Gene)": "INTERACTION (Cond x Gene)"
        }
        anova_table.index = [new_indices.get(x, x) for x in anova_table.index]
        
        # Rimuoviamo l'intercetta per focus sugli effetti sperimentali
        if "Intercept" in anova_table.index:
            anova_table = anova_table.drop("Intercept")

        # Formattazione estetica
        def color_significant(val):
            return 'color: red; font-weight: bold' if val < 0.05 else 'color: black'

        print(f"\n>>> AREA: {area}")
        
        # Visualizzazione della tabella ANOVA
        styled_table = anova_table.style.map(color_significant, subset=['pvalue'])
        display(styled_table)

        # Commento analitico automatico
        try:
            p_cond = anova_table.loc["CONDITION (Between: A, O, C)", "pvalue"]
            p_inter = anova_table.loc["INTERACTION (Cond x Gene)", "pvalue"]
            
            if p_cond < 0.05:
                print(f"\033[91m[!] EFFETTO CONDIZIONE SIGNIFICATIVO: Il trattamento (A o O) ha spostato l'espressione genica in {area}.\033[0m")
            
            if p_inter < 0.05:
                print(f"\033[94m[*] INTERAZIONE SIGNIFICATIVA: I geni in {area} non rispondono tutti allo stesso modo alla condizione.\033[0m")
            elif p_cond >= 0.05 and p_inter >= 0.05:
                print(f"[-] Nessun effetto rilevante dei fattori principali in quest'area.")
                
        except KeyError:
            pass

    except Exception as e:
        print(f"Errore nell'elaborazione dell'area {area}: {e}")
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
import pingouin as pg
from IPython.display import display

# 1. Filtriamo per le condizioni principali (A, O, C)
# Assicurati che pcr_clean sia il DataFrame con i dati puliti
df_lmm = pcr_clean[pcr_clean['Condition'].isin(['A', 'O', 'C'])].copy()

genes = df_lmm['Gene'].unique()

print(f"\033[1m--- ANALISI MODELLI MISTI: INTERAZIONE CONDITION * AREA ---\033[0m")
print("Modello: DCt ~ Condition * Area | Effetto Random: SubID\n")

for g in genes:
    print(f"\n" + "="*65)
    print(f"GENE: {g}")
    print("="*65)
    
    gene_data = df_lmm[df_lmm['Gene'] == g].copy()
    
    try:
        # 1. COSTRUZIONE DEL MODELLO (LMM)
        # In R: lmer(DCt ~ Condition * Area + (1|SubID))
        model = smf.mixedlm("DCt ~ C(Condition, Treatment('C')) * Area", 
                            gene_data, groups=gene_data["SubID"])
        result = model.fit()
        
        # 2. TABELLA ANOVA (Type III Wald Test)
        # Questo corrisponde a print(anova(model)) in R
        wald_test = result.wald_test_terms(scalar=False)
        anova_table = wald_test.table
        
        # Rinominiamo gli indici per chiarezza
        new_indices = {
            "C(Condition, Treatment('C'))": "Condition",
            "Area": "Area",
            "C(Condition, Treatment('C')):Area": "Condition * Area"
        }
        anova_table.index = [new_indices.get(x, x) for x in anova_table.index]
        
        # Rimuoviamo l'intercetta e visualizziamo
        if "Intercept" in anova_table.index:
            anova_table = anova_table.drop("Intercept")
            
        print("\n[ANOVA Table - Wald Test]")
        display(anova_table.style.map(lambda x: 'color: red; font-weight: bold' if x < 0.05 else 'color: black', subset=['pvalue']))

        # 3. TEST POST-HOC (Confronti a coppie per Area)
        # Corrisponde a emmeans(model, specs = pairwise ~ Condition | Area)
        print(f"\n[Post-hoc: Condition per Area per {g}]")
        
        posthoc_results = []
        for area in gene_data['Area'].unique():
            area_subset = gene_data[gene_data['Area'] == area]
            
            if area_subset['Condition'].nunique() > 1:
                # Eseguiamo i test pairwise (Tukey o FDR)
                ph = pg.pairwise_tests(data=area_subset, dv='DCt', between='Condition', padjust='fdr')
                ph['Area'] = area
                posthoc_results.append(ph)
        
        # Uniamo e mostriamo i post-hoc significativi o i trend
        if posthoc_results:
            all_ph = pd.concat(posthoc_results, ignore_index=True)
            # Cerchiamo la colonna p-value corretta
            p_col = [c for c in all_ph.columns if 'p-' in c or 'pval' in c or 'p-val' in c][0]
            
            # Filtriamo per mostrare risultati rilevanti (p < 0.10)
            sig_ph = all_ph[all_ph[p_col] < 0.10].copy()
            if not sig_ph.empty:
                display(sig_ph[['Area', 'A', 'B', 'stat', p_col]].sort_values(by=['Area', p_col]))
            else:
                print("Nessun confronto post-hoc significativo (p < 0.10) in nessuna area.")

    except Exception as e:
        print(f"Errore nell'analisi del gene {g}: {e}")
        
```

```python
import warnings
import statsmodels.formula.api as smf
import pandas as pd
from IPython.display import display

print(f"\033[1m--- ANALISI POST-HOC MIRATA: O_i vs CONTROL --- \033[0m")

for area in aree_ordine:
    df_area = df_sep[df_sep['Area'] == area].copy()
    
    # Modello Mixed (DCt normalizzato)
    model = smf.mixedlm("DCt_norm ~ C(Condition, Treatment('C')) * C(Gene)", 
                        df_area, groups=df_area["SubID"])
    result = model.fit()
    
    # Estraiamo la tabella dei coefficienti
    # Il coefficiente "C(Condition...)[T.O_i]" testa direttamente O_i contro il Controllo (C)
    summary_df = result.summary().tables[1]
    summary_df['P>|z|'] = pd.to_numeric(summary_df['P>|z|'])
    
    print(f"\n>>> AREA: {area}")
    
    # Focalizziamoci sui confronti chiave
    confronti_chiave = [
        f"C(Condition, Treatment('C'))[T.O_i]",
        f"C(Condition, Treatment('C'))[T.O_ni]"
    ]
    
    # Filtriamo la tabella per mostrare solo questi
    present_comparisons = [c for c in confronti_chiave if c in summary_df.index]
    subset_df = summary_df.loc[present_comparisons].copy()
    
    # Rinominiamo per leggibilità
    subset_df.index = ["Imitatori (O_i) vs Controllo", "Non-Imitatori (O_ni) vs Controllo"]

    # Formattazione con map (ex applymap)
    styled_subset = subset_df.style.map(
        lambda val: 'color: red; font-weight: bold' if val < 0.05 else 'color: black', 
        subset=['P>|z|']
    )
    
    display(styled_subset)
    
    # Logica per commento rapido
    p_oi = subset_df.loc["Imitatori (O_i) vs Controllo", "P>|z|"]
    if p_oi < 0.05:
        print(f"\033[91m[SUCCESS] Trovata differenza significativa per gli Imitatori in {area}!\033[0m")
    elif p_oi < 0.10:
        print(f"\033[93m[TREND] C'è un trend (p={p_oi:.3f}) per gli Imitatori in {area}.\033[0m")
```

```python
import warnings
import statsmodels.formula.api as smf
import pandas as pd
from IPython.display import display

# Zittiamo i FutureWarning
warnings.simplefilter(action='ignore', category=FutureWarning)

print(f"\033[1m--- ANALISI POST-HOC MIRATA: ATTORE (A) vs CONTROL (C) --- \033[0m")
print("Analisi sui dati normalizzati (Z-score) - Modello Mixed\n")

for area in aree_ordine:
    df_area = df_sep[df_sep['Area'] == area].copy()
    
    # Modello Mixed (DCt normalizzato)
    # Fissiamo 'C' come riferimento per il confronto diretto
    model = smf.mixedlm("DCt_norm ~ C(Condition, Treatment('C')) * C(Gene)", 
                        df_area, groups=df_area["SubID"])
    result = model.fit()
    
    # Estraiamo la tabella dei coefficienti
    summary_df = result.summary().tables[1]
    summary_df['P>|z|'] = pd.to_numeric(summary_df['P>|z|'])
    
    print(f"\n>>> AREA: {area}")
    
    # Identifichiamo il confronto specifico per l'Attore
    confronto_attore = f"C(Condition, Treatment('C'))[T.A]"
    
    if confronto_attore in summary_df.index:
        subset_df = summary_df.loc[[confronto_attore]].copy()
        
        # Rinominiamo per chiarezza
        subset_df.index = ["Attore (A) vs Controllo (C)"]

        # Formattazione: Rosso se p < 0.05, Arancione se p < 0.10
        def highlight_p(val):
            if val < 0.05: return 'color: red; font-weight: bold'
            if val < 0.10: return 'color: orange; font-weight: bold'
            return 'color: black'

        styled_subset = subset_df.style.map(highlight_p, subset=['P>|z|'])
        display(styled_subset)
        
        # Logica per commento testuale
        p_a = subset_df.loc["Attore (A) vs Controllo (C)", "P>|z|"]
        coef_a = subset_df.loc["Attore (A) vs Controllo (C)", "Coef."]
        
        if p_a < 0.05:
            direzione = "upregulation (più espressione)" if coef_a < 0 else "downregulation (meno espressione)"
            print(f"\033[91m[SUCCESS] Differenza significativa per l'Attore in {area}! ({direzione})\033[0m")
        elif p_a < 0.10:
            print(f"\033[93m[TREND] C'è un trend (p={p_a:.3f}) per l'Attore in {area}.\033[0m")
        else:
            print(f"[-] Nessuna differenza rilevante per l'Attore in {area}.")
    else:
        print(f"[!] Errore: Gruppo 'A' non trovato nei dati per l'area {area}.")

print("\n--- Fine Analisi ---")
```

```python
# 1. Creiamo la macro-categoria "Observer"
df_sep['Condition_Collapsed'] = df_sep['Condition'].replace({'O_i': 'Observer', 'O_ni': 'Observer'})

print(f"\033[1m--- ANALISI MACRO-GRUPPO: OBSERVER (O_i+O_ni) vs CONTROL --- \033[0m")

for area in ["Wulst", "IMM", "MSt", "Nidopallium", "Arcopallium"]:
    df_area = df_sep[df_sep['Area'] == area].copy()
    
    # Modello con la nuova condizione collassata
    # Usiamo 'C' come riferimento
    model = smf.mixedlm("DCt_norm ~ C(Condition_Collapsed, Treatment('C')) + C(Gene)", 
                        df_area, groups=df_area["SubID"])
    result = model.fit()
    
    summary_df = result.summary().tables[1]
    summary_df['P>|z|'] = pd.to_numeric(summary_df['P>|z|'])
    
    print(f"\n>>> AREA: {area}")
    
    # Selezioniamo solo il confronto Observer vs Controllo
    idx_obs = "C(Condition_Collapsed, Treatment('C'))[T.Observer]"
    if idx_obs in summary_df.index:
        res_obs = summary_df.loc[[idx_obs]].copy()
        res_obs.index = ["Tutti gli Osservatori vs Controllo"]
        
        # Coloriamo i p-value
        display(res_obs.style.map(
            lambda val: 'color: red; font-weight: bold' if val < 0.05 else ('color: orange; font-weight: bold' if val < 0.10 else 'color: black'), 
            subset=['P>|z|']
        ))
```

```python
#2. ANOVA One-Way o Kruskal-Wallis (Area per Area, Gene per Gene) e Fisher
#È l'approccio classico e più sicuro per una tesi.
#Fai un test separato per ogni area e ogni gene (quello che stavamo iniziando a fare con gli asterischi).

import pandas as pd
from scipy import stats
from IPython.display import display

# Creiamo le due varianti della condizione
df_sep['Condition_Collapsed'] = df_sep['Condition'].replace({'O_i': 'Observer', 'O_ni': 'Observer'})
genes_ordine = ['cFOS', 'ARC', 'EGR1']
results_list = []

print(f"\033[1m--- ANALISI GRANULARE: GENE PER GENE, AREA PER AREA ---\033[0m\n")

for area in aree_ordine:
    for gene in genes_ordine:
        # Filtriamo i dati
        subset = df_sep[(df_sep['Area'] == area) & (df_sep['Gene'] == gene)]
        
        # 1. TEST PER GRUPPI SEPARATI (A, O_i, O_ni, C)
        groups = [subset[subset['Condition'] == c]['DCt'].values for c in ['A', 'O_i', 'O_ni', 'C']]
        # Rimuoviamo array vuoti se presenti
        groups = [g for g in groups if len(g) > 0]
        
        f_stat, p_anova = stats.f_oneway(*groups) if len(groups) > 1 else (0, 1)
        h_stat, p_kruskal = stats.kruskal(*groups) if len(groups) > 1 else (0, 1)
        
        # 2. TEST PER OBSERVER INSIEME (A, Observer, C)
        groups_coll = [subset[subset['Condition_Collapsed'] == c]['DCt'].values for c in ['A', 'Observer', 'C']]
        groups_coll = [g for g in groups_coll if len(g) > 0]
        
        f_stat_c, p_anova_c = stats.f_oneway(*groups_coll) if len(groups_coll) > 1 else (0, 1) #fisher 
        
        # Salviamo i risultati
        results_list.append({
            'Area': area,
            'Gene': gene,
            'p_ANOVA (4 grp)': p_anova,
            'p_Kruskal (4 grp)': p_kruskal,
            'p_ANOVA (Obs Unito)': p_anova_c
        })

# Creiamo il DataFrame riassuntivo
df_stats = pd.DataFrame(results_list)

# Funzione per colorare e mettere in grassetto i p-value significativi
def highlight_stats(val):
    if val < 0.05:
        return 'color: red; font-weight: bold'
    elif val < 0.10:
        return 'color: orange'
    return 'color: black'

# Visualizzazione tabellare
styled_results = df_stats.style.map(highlight_stats, subset=['p_ANOVA (4 grp)', 'p_Kruskal (4 grp)', 'p_ANOVA (Obs Unito)'])
display(styled_results)
```

```python
#confronti a coppie post hoc

from statsmodels.stats.multicomp import pairwise_tukeyhsd

def run_posthoc(area, gene, col_condizione):
    subset = df_sep[(df_sep['Area'] == area) & (df_sep['Gene'] == gene)]
    print(f"\n--- Post-hoc per {area} - {gene} (usando {col_condizione}) ---")
    
    # Eseguiamo il test di Tukey
    tukey = pairwise_tukeyhsd(endog=subset['DCt'], groups=subset[col_condizione], alpha=0.05)
    print(tukey)

# 1. Analisi per Arcopallium ARC (Osservatori Uniti)
run_posthoc('Arcopallium', 'ARC', 'Condition_Collapsed')

# 2. Analisi per IMM ARC (Gruppi separati)
run_posthoc('IMM', 'ARC', 'Condition')
```

```python
#3. Analisi Multivariata (MANOVA)
#Invece di normalizzare e unire, tratti i 3 geni come variabili dipendenti multiple correlate.
#Vantaggio: Ti dice se la "combinazione" dei tre geni differisce tra i gruppi, tenendo conto della loro correlazione (che hai già visto nelle heatmap).

from statsmodels.multivariate.manova import MANOVA

print(f"\033[1m--- ANALISI MULTIVARIATA (MANOVA) PER AREA ---\033[0m")
print("Variabili dipendenti: [ARC, EGR1, cFOS] | Fattore: Condition\n")

for area in aree_ordine:
    # Prepariamo i dati: mettiamo i geni in colonne separate
    df_area = df_clean[df_clean['Area'] == area].pivot_table(
        index=['SubID', 'Condition'], 
        columns='Gene', 
        values='DCt'
    ).reset_index()
    
    # Rimuoviamo eventuali righe con dati mancanti (la MANOVA non li tollera)
    df_area = df_area.dropna()
    
    if len(df_area) > 10:  # Serve un numero minimo di campioni
        # Specifichiamo la formula: i 3 geni sono le variabili dipendenti
        maov = MANOVA.from_formula('ARC + EGR1 + cFOS ~ Condition', data=df_area)
        
        # Eseguiamo il test
        res = maov.mv_test()
        
        # Estraiamo il p-value di Wilks' Lambda per la Condition
        p_val = res.results['Condition']['stat'].loc["Wilks' lambda", "Pr > F"]
        
        # Formattazione colore
        color = "\033[91m" if p_val < 0.05 else ("\033[93m" if p_val < 0.10 else "\033[0m")
        print(f"Area: {area:<15} | Wilks' Lambda p-value: {color}{p_val:.4f}\033[0m")
    else:
        print(f"Area: {area:<15} | Dati insufficienti per MANOVA")
```

```python
from scipy.stats import pearsonr
import pandas as pd

results_corr = []

# Cicliamo per area e condizione per calcolare la significatività di ogni coppia
for area in aree_ordine:
    for cond in df_clean['Condition'].unique():
        # Filtriamo i dati
        subset = df_clean[(df_clean['Area'] == area) & (df_clean['Condition'] == cond)].pivot_table(
            index='SubID', columns='Gene', values='DCt'
        ).dropna()
        
        # Coppie di geni da confrontare
        coppie = [('ARC', 'EGR1'), ('ARC', 'cFOS'), ('EGR1', 'cFOS')]
        
        for g1, g2 in coppie:
            if g1 in subset.columns and g2 in subset.columns and len(subset) > 2:
                r_val, p_val = pearsonr(subset[g1], subset[g2])
                results_corr.append({
                    'Area': area,
                    'Condizione': cond,
                    'Coppia': f"{g1}-{g2}",
                    'R_Pearson': r_val,
                    'p-value': p_val,
                    'N': len(subset)
                })

# Creiamo il DataFrame
df_corr_stats = pd.DataFrame(results_corr)

# Applichiamo la formattazione per evidenziare le correlazioni significative
def highlight_corr(val):
    if val < 0.05:
        return 'background-color: #ffcccc; font-weight: bold' # Rosso chiaro per significativo
    elif val < 0.10:
        return 'background-color: #fff2cc' # Giallo chiaro per trend
    return ''

styled_corr = df_corr_stats.style.map(highlight_corr, subset=['p-value'])

print(f"\033[1m--- TABELLA SIGNIFICATIVITÀ DELLE CORRELAZIONI ---\033[0m")
display(styled_corr)
```

```python
# parte di mirko

import pandas as pd
import numpy as np
import pingouin as pg
import statsmodels.formula.api as smf
from statsmodels.stats.multitest import multipletests

# ========================================================
# 1. CARICAMENTO E PRE-PROCESSING
# ========================================================
filename = "imitation_data_complete.xlsx"

# Import dei fogli specifici
data_behav = pd.read_excel(filename, sheet_name="data_behav")
data_pcr = pd.read_excel(filename, sheet_name="data_pcr")

# Pulizia e conversione in fattori (category in Python)
pcr_clean = data_pcr[['SubID', 'Condition', 'Area', 'Gene', 'DCt']].copy()
for col in ['SubID', 'Condition', 'Area', 'Gene']:
    pcr_clean[col] = pcr_clean[col].astype('category')

# Summary Tables
print("\n--- Summary for Behavioral Data ---")
print(data_behav.groupby('Condition')['SubID'].nunique())

print("\n--- Summary for PCR Data ---")
print(pcr_clean.groupby('Condition')['SubID'].nunique())

# ========================================================
# 2. VERIFICA ASSUNZIONI ANOVA
# ========================================================
print("\n" + "="*56)
print("           ANOVA ASSUMPTIONS SUMMARY REPORT")
print("="*56)

# --- 1. OUTLIERS (Z-score > 3 come extreme) ---
pcr_clean['z_score'] = pcr_clean.groupby(['Gene', 'Condition', 'Area'])['DCt'].transform(lambda x: (x - x.mean()) / x.std())
outliers = pcr_clean[pcr_clean['z_score'].abs() > 3]

if outliers.empty:
    print("\n1. OUTLIERS: No extreme outliers detected. (Perfect!)")
else:
    print("\n1. OUTLIERS: Warning! Extreme outliers found:")
    print(outliers[['Gene', 'Condition', 'Area', 'SubID', 'DCt']])

# --- 2. NORMALITY (Shapiro-Wilk) ---
# Usiamo groupby().apply() per aggirare il limite della lista in pg.normality
normality = pcr_clean.groupby(['Gene', 'Condition', 'Area'], observed=True).apply(
    lambda x: pg.normality(x['DCt']) if len(x) >= 3 else pd.DataFrame({'normal': [None], 'p': [np.nan]})
).reset_index()

# Pulizia dei risultati (rimuoviamo i livelli extra creati dal groupby)
normality = normality.drop(columns=['level_3'] if 'level_3' in normality.columns else [])

failed_norm = normality[normality['normal'] == False]

print("\n2. NORMALITY CHECK (Shapiro-Wilk):")
if failed_norm.empty:
    print("   All groups are normally distributed (p > 0.05).")
else:
    print("   The following groups violated normality:")
    print(failed_norm[['Gene', 'Condition', 'Area', 'pval']])

# --- 3. HOMOGENEITY OF VARIANCE (Levene's) ---
# Anche qui usiamo un ciclo manuale per sicurezza
levene_list = []
for (gene, area), group in pcr_clean.groupby(['Gene', 'Area'], observed=True):
    # Eseguiamo Levene solo se abbiamo almeno 2 gruppi con dati
    if group['Condition'].nunique() > 1:
        lev = pg.homoscedasticity(group, dv='DCt', group='Condition')
        if not lev['equal_var'].iloc[0]:
            levene_list.append({'Gene': gene, 'Area': area, 'p': lev['pval'].iloc[0]})

print("\n3. HOMOGENEITY OF VARIANCE (Levene's):")
if not levene_list:
    print("   Variance is equal across groups for all areas (p > 0.05).")
else:
    print(pd.DataFrame(levene_list))

# ========================================================
# 3. LINEAR MIXED MODELS (LMM) PER OGNI GENE
# ========================================================
genes = pcr_clean['Gene'].unique()

for g in genes:
    print(f"\n--- Running Mixed Model for Gene: {g} ---")
    gene_data = pcr_clean[pcr_clean['Gene'] == g].copy()
    
    # Modello: DCt ~ Condition * Area + (1|SubID)
    # Nota: MixedLM in Python usa 'groups' per l'effetto random
    model = smf.mixedlm("DCt ~ Condition * Area", gene_data, groups=gene_data["SubID"]).fit()
    
    print("\nANOVA-like Table (Fixed Effects):")
    print(model.summary().tables[1]) # Mostra coefficienti e p-values
    
# Post-hoc (Pairwise comparisons per ogni Area)
    print(f"\nPost-hoc comparisons (FDR adjusted) for {g}:")
    for area in gene_data['Area'].unique():
        area_subset = gene_data[gene_data['Area'] == area]
        
        # Procediamo solo se abbiamo almeno 2 soggetti per almeno 2 condizioni diverse
        if area_subset['Condition'].nunique() > 1:
            ph = pg.pairwise_tests(data=area_subset, dv='DCt', between='Condition', padjust='fdr')
            
            # Controllo dinamico delle colonne esistenti
            cols_to_print = ['A', 'B']
            # Cerca i nomi delle colonne p-value (possono variare tra versioni)
            for p_col in ['p-unc', 'p-corr', 'pval', 'p-adjust', 'hedges']:
                if p_col in ph.columns:
                    cols_to_print.append(p_col)
            
            print(f"\nArea: {area}")
            if not ph.empty:
                print(ph[cols_to_print])
            else:
                print("Nessun confronto possibile (dati insufficienti).")
```
