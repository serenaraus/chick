## Exercise 4 — Count files by extension

Given a folder, count how many files end with:

- `.py`
- `.ipynb`
- `.md`

Print the three counts.

```python
from pathlib import Path

folder = Path(r"C:\Users\seren\Documents\GitHub\chick\QuaQua")

py_count = 0
ipynb_count = 0
md_count = 0

for path in folder.glob("*"):
    if path.suffix == ".py":
        py_count += 1
    elif path.suffix == ".ipynb":
        ipynb_count += 1
    elif path.suffix == ".md":
        md_count += 1

print("py:", py_count)
print("ipynb:", ipynb_count)
print("md:", md_count)
```

## Exercise 10 — Organize files by prefix

Suppose a folder contains files with names such as:

- `sub01_session1_data.csv`
- `sub01_session2_data.csv`
- `sub02_session1_data.csv`

Write code that moves each file into a folder structure like:

- `sub01/session1/`
- `sub01/session2/`
- `sub02/session1/`

Use the filename (without extension) and split it on `"_"`. 

```python
from pathlib import Path

data_folder = Path(r"C:\Users\seren\Documents\GitHub\chick\QuaQua")
data_folder.mkdir(parents=True, exist_ok=True)
soggetti = ["lago", "albero", "montagna"]
sessioni = ["session1", "session2", "session3"]

for s in soggetti:
    for sess in sessioni:
        filename = f"{s}_{sess}_data.csv"
        (data_folder / filename).touch()

print("File creati con successo.")
```

```python
from pathlib import Path

data_folder = Path(r"C:\Users\seren\Documents\GitHub\chick\QuaQua")
new_root = Path(r"C:\Users\seren\Documents\GitHub\chick\QuaQua\organized")
new_root.mkdir(exist_ok=True)

for file in data_folder.glob("*.csv"):
    subject, session, _ = file.stem.split("_")
    target_folder = new_root / subject / session
    target_folder.mkdir(exist_ok=True, parents=True)
    file.replace(target_folder / file.name)
    print(f"Organizzato {file.name} in {subject} > {session}")
```

## Exercise 11 — Build a file inventory

Given a folder, create a list of dictionaries where each dictionary contains:

- `"name"`
- `"suffix"`
- `"is_dir"`
- `"parent"`

for every path in the folder.

Then print the resulting list.

```python
from pathlib import Path

folder = Path(r"C:\Users\seren\Documents\GitHub\chick\QuaQua")
inventory = []

for path in folder.rglob("*"):
    info = {
        "name": path.name,
        "suffix": path.suffix,
        "is_dir": path.is_dir(),
        "parent": path.parent.name
    }
    inventory.append(info)
    
from pprint import pprint #pretty print
pprint(inventory)
```

## Exercise 12 —  Only files, sorted by suffix

Given a folder:

1. collect only the files, not the subfolders
2. sort them by suffix
3. print their names

```python
from pathlib import Path

folder = Path(r"C:\Users\seren\Documents\GitHub\chick\QuaQua")

files_only = []

for path in folder.glob("*"):
    if not path.is_dir():
        files_only.append(path)

for path in sorted(files_only, key=lambda p: p.suffix):
    print(path.name)
```
