## Exercise 1 — Count letters, digits, and symbols

Complete the code below so that it counts:
- alphabetic characters (`character.isalpha`),
- digits(`character.isdigit`),
- all remaining symbols

in the string `stimulus_code`.

```python
stimulus_code = "EEG_trial12@bad#"

# set the initial values for the counters
word_characters_count = 0
digits_count = 0
symbols_count = 0

# iterate over the string
for ch in stimulus_code:
    if ch.isalpha():    # check if a letter
        word_characters_count += 1
    elif ch.isdigit():  # check if a digit
        digits_count += 1
    else:               # everything else is a symbol
        symbols_count += 1

counts = {
    "Word Characters": word_characters_count,
    "Digits": digits_count,
    "Symbols": symbols_count
}
print(counts)
```

## Exercise 2 — Fibonacci sequence

Write code to generate the first 10 values of the Fibonacci series (a series of numbers where each number (Fibonacci number) is the sum of the two preceding ones, typically starting from 0 and 1).

```python
fibonacci_series = [0, 1]

while len(fibonacci_series) < 10:
    fibonacci_series.append(fibonacci_series[-1] + fibonacci_series[-2])

print(fibonacci_series)
```

## Exercise 3 — Top 3 subjects by accuracy

You performed a learning task and obtained subject accuracies.

```python
accuracy_scores = {
    'id1': 0.27, 'id2': 0.75, 'id3': 0.61, 'id4': 0.05, 'id5': 0.40,
    'id6': 0.67, 'id7': 0.69, 'id8': 0.52, 'id9': 0.70, 'id10': 0.30
}

# store the top 3 accuracy values as a list
max_accs = sorted(accuracy_scores.values(), reverse=True)[:3]

# create an empty list that will hold ids of participants with the highest accuracy
max_ids = []

for subject_id, acc in accuracy_scores.items():
    if acc in max_accs:
        max_ids.append(subject_id)

print(max_ids)
```

## Exercise 4 — Exclude impossible reaction times

Some RT values are invalid because they are zero or negative.  
Create a new list `valid_rts` containing only values strictly greater than zero.

First do it with a `for` loop, then do it with a list comprehension.

```python
rts = [0.61, 0.72, 0, 0.55, -0.3, 0.81, 0.49]

# Version 1: for loop
valid_rts = []

for rt in rts:
    if rt > 0:
        valid_rts.append(rt)

print(valid_rts)

# Version 2: list comprehension
valid_rts_comp = [rt for rt in rts if rt > 0]

print(valid_rts_comp)
```

## Exercise 5 — Variance from standard deviation

Following the lecture example, compute variances from a list of standard deviations.  
If the standard deviation is `0`, insert `"invalid"` instead of squaring it.

Do this with a list comprehension using `if ... else`.

```python
stdev = [2, 4, 0, 1.5, 2, 4]

variances = [val**2 if val != 0 else "invalid" for val in stdev]

print(variances)
```

## Exercise 7 — Categorize scores with `if/elif/else`

Given a list of exam scores, assign a label to each one:
- `"low"` if score < 18
- `"pass"` if score is between 18 and 24 inclusive
- `"good"` if score is between 25 and 28 inclusive
- `"excellent"` if score > 28

Store the labels in a new list.

```python
scores = [12, 18, 21, 25, 28, 30, 17, 24]

labels = []

for score in scores:
    if score < 18:
        labels.append("low")
    elif score <= 24:
        labels.append("pass")
    elif score <= 28:
        labels.append("good")
    else:
        labels.append("excellent")

print(labels)
```

## Exercise 8 — Best coding option

Two students wrote code to create a list of squared positive values.

Tasks:
1. Run both versions.
2. Check that they produce the same output.
3. Decide which one is more Pythonic **for this case**.
4. Write one sentence explaining your choice.

```python
# Option A
numbers = [-3, -1, 0, 2, 4]
result = []
for x in numbers:
    if x > 0:
        result.append(x ** 2)
print(result)

# Option B
numbers = [-3, -1, 0, 2, 4]
result = [x ** 2 for x in numbers if x > 0]
print(result)

# Option B is more "Pythonic" because it uses a list comprehension to filter and transform in only a single easy to read line of code
```

## Exercise 9 — Use `enumerate()` for indexed reporting

You collected amplitudes from five channels. Print a message like
`Channel 0 has amplitude 12.4`
for each value, using `enumerate()`.

Then create a list `high_amplitude_channels` containing the indices of channels with amplitude greater than 11.

```python
amplitudes = [12.4, 8.1, 15.2, 9.7, 11.0]

for i, amp in enumerate(amplitudes):
    print(f"Channel {i} has amplitude of {amp} µV")

# list of indices with amplitude > 11
high_amplitude_channels = [i for i, amp in enumerate(amplitudes) if amp > 11]
print(f"Channels {high_amplitude_channels} have amplitude > 11µV")
```

## Exercise 10 — Use `zip()` to combine subject IDs and scores

You have two parallel lists.

Tasks:
1. print each subject and score using `zip()`
2. create a dictionary `score_by_subject`
3. create a list containing only the IDs of subjects with score > 20

```python
subject_ids = ["s01", "s02", "s03", "s04"]
memory_scores = [21, 18, 25, 19]

for subject_id, score in zip(subject_ids, memory_scores):
    print(subject_id, "->", score)

# dictionary score_by_subject
score_by_subject = dict(zip(subject_ids, memory_scores))

# list containing only the IDs of subjects  with score > 20
high_scorers = [sid for sid, score in zip(subject_ids, memory_scores) if score > 20]

print(score_by_subject)
print(high_scorers)
```

## Exercise 11 — Stop at the first invalid value with `break`

You are scanning a list of standard deviations.  
As soon as you encounter a zero, stop the loop completely and keep only the variances computed so far.

```python
stdev = [2, 4, 1.5, 0, 3, 2]
variances = []

for val in stdev:
    if val == 0:
        break
    variances.append(val**2)

print(variances)
```

## Exercise 12 — Skip invalid values with `continue`

Using the same list, compute variances again, but this time skip only the zero and continue processing the remaining values.

```python
stdev = [2, 4, 1.5, 0, 3, 2]
variances = []

for val in stdev:
    if val == 0:
        continue
    variances.append(val**2)

print(variances)
```

## Exercise 13 — Nested loop over subjects and trials

You have trial accuracies for three subjects.

Tasks:
1. compute the mean accuracy for each subject
2. store the results in a list called `mean_accuracies`
3. print the subject number and its mean accuracy

```python
all_scores = [
    [1, 0, 1, 1],
    [0, 0, 1, 1],
    [1, 1, 1, 0]
]

mean_accuracies = []

for i, subject_scores in enumerate(all_scores):
    mean_acc = sum(subject_scores) / len(subject_scores)
    mean_accuracies.append(mean_acc)
    print(f"Subject {i}: {mean_acc}")
```

## Exercise 14 — Build a dictionary from two lists

Create a dictionary mapping each brain region to its average activity.

First use a `for` loop with `zip()`.  
Then recreate the same dictionary with a dictionary comprehension.

```python
regions = ["frontal", "parietal", "temporal", "occipital"]
activity = [0.82, 0.76, 0.91, 0.68]

# Version 1
activity_by_region = {}
for region, value in zip(regions, activity):
    activity_by_region[region] = value
print(activity_by_region)

# Version 2
activity_by_region_comp = {region: value for region, value in zip(regions, activity)}
print(activity_by_region_comp)
```

## Exercise 15 — Running sum with `while`

Using a `while` loop, compute the cumulative sum of the numbers in the list below until the sum becomes greater than or equal to 20.  
Then stop and print:
- the final sum
- the index at which the loop stopped

```python
values = [3, 5, 4, 8, 2, 7]
total = 0
i = 0

while i < len(values) and total < 20:
    total += values[i]
    i += 1
    print(total)

print("Final total:", total)
print("Stopped at index:", i)
```

Create a new dictionary from `rt_by_subject` such that:

## Optional challenge 1 — Prime numbers up to 30

Create a list of all prime numbers from 2 to 30. You may use nested loops and a boolean flag.

```python
primes = []

for n in range(2, 31):
    is_prime = True
    for divisor in range(2, n):
        if n % divisor == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(n)

print(primes)
```

## Optional challenge 2 — Longest streak of correct responses

Given a list of binary responses, compute the longest consecutive run of `1`s.

```python
responses = [1, 1, 0, 1, 1, 1, 0, 1, 1]

longest_streak = 0
current_streak = 0

for response in responses:
    if response == 1:
        current_streak += 1
        if current_streak > longest_streak:
            longest_streak = current_streak
    else:
        current_streak = 0

print(longest_streak)
```
