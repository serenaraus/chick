````python
# Let's define a class
# 1. What is the class trying to represent? Different data from a research, such as participants, reaction times and experimental records.
# 2. What attributes are set in `__init__()`? 
# 3. Which methods are available? 
# 4. Which methods return values, and which modify the object?  
# 5. Are there properties?  
# 6. Is the class inheriting from another class?  
# 7. Are there class attributes shared across instances?

class ParticipantData:
    """Represent data from an experimental participant."""

    def __init__(self, participant_id, age, condition, base_path="/path/to/data"):
        # Store core participant information
        self.participant_id = participant_id
        self.age = age
        self.condition = condition

        # Base folder where data files are assumed to live
        self.base_path = base_path

        # Fixed experiment label for all instances of this class
        self.experiment_name = "Experiment 1"

    def get_data_path(self):
        # Build the expected filename for this participant
        return f"{self.base_path}/subject_{self.participant_id}_{self.condition}.csv"

    def get_exp_metadata(self):
        # Return participant-related metadata as a dictionary
        return {
            "participant_id": self.participant_id,
            "age": self.age,
            "condition": self.condition,
            "experiment_name": self.experiment_name,
        }

    def get_exp_data(self):
        # Compute the path to the participant's data file
        data_path = self.get_data_path()

        # Simulate loading data from disk
        print(f"Loading data from {data_path} ...")

        # Placeholder return value standing in for real loaded data
        return [1, 2, 3]

    def __repr__(self):
        # Developer-friendly string representation of the object
        return (
            f"ParticipantData(participant_id={self.participant_id!r}, "
            f"age={self.age!r}, condition={self.condition!r})"
        )


class IntrospectiveClass:
    def __init__(self, attribute=0):
        # Store one instance attribute, with default value 0
        self.my_attribute = attribute

    def __str__(self):
        # Human-readable string representation of the object
        return f"IntrospectiveClass(my_attribute={self.my_attribute})"

    def introspect(self):
        # Print information about the current object
        print(f"Who am I? {self}")
        print(f"My attribute value is: {self.my_attribute}")


class ReactionTimesAnalyzer:
    """Class to analyze reaction time data."""

    def __init__(self, reaction_times):
        # Store a copy so external changes to the original list
        # do not affect the object accidentally
        self.reaction_times = reaction_times.copy()

    def get_average_rt(self):
        # Compute and return the mean reaction time
        return sum(self.reaction_times) / len(self.reaction_times)

    def append_new_trial(self, reaction_time):
        # Add one new reaction time to the stored data
        self.reaction_times.append(reaction_time)


class Student:
    """A very small class representing a student."""

    def __init__(self, name, age, enrollment="CIMeC PhD"):
        # Store basic student information
        self.name = name
        self.age = age

        # Optional argument with a default value
        self.enrollment = enrollment

    def introduce(self):
        # Return a short self-introduction string
        return f"My name is {self.name}, I am {self.age}, and I am enrolled in {self.enrollment}."


class BehavioralExperiment:
    def __init__(self, reaction_times):
        # Store a copy of the reaction times list
        self.reaction_times = reaction_times.copy()

    def append_new_trial_rt(self, reaction_time):
        # Modify the object by adding a new trial
        self.reaction_times.append(reaction_time)

    @property
    def mean_rt(self):
        # Property: behaves like an attribute but computes on demand

        # Return None if there are no reaction times
        if len(self.reaction_times) == 0:
            return None

        # Otherwise return the mean reaction time
        return sum(self.reaction_times) / len(self.reaction_times)


class SubjectRecord:
    def __init__(self, subject_id, reaction_times):
        # Leading underscore: internal/protected-by-convention attribute
        self._subject_id = subject_id

        # Store a copy of the reaction times
        self.reaction_times = reaction_times.copy()

    @property
    def subject_id(self):
        # Read-only property exposing the internal subject ID
        return self._subject_id


class Person:
    def __init__(self, name, age):
        # Public attribute
        self.name = name

        # Internal storage for age
        self._age = age

    @property
    def age(self):
        # Getter: allows reading age like an attribute
        return self._age

    @age.setter
    def age(self, value):
        # Setter: validates before updating age
        if value < 0:
            raise ValueError("Age cannot be negative.")
        self._age = value


class DemoClass:
    # Class attribute: shared by all instances of the class
    class_attribute = []

    def __init__(self):
        # Instance attribute: each object gets its own list
        self.instance_attribute = []


class PersonBase:
    def __init__(self, name, age):
        # Store shared person information
        self.name = name
        self.age = age

    def introduce(self):
        # Print a basic introduction
        print(f"My name is {self.name} and I am {self.age} years old.")


class StudentPerson(PersonBase):
    def __init__(self, name, age, student_id):
        # Reuse initialization from the parent class
        super().__init__(name, age)

        # Add the attribute specific to students
        self.student_id = student_id

    def introduce(self):
        # First call the parent version of introduce()
        super().introduce()

        # Then extend it with student-specific information
        print(f"My student ID is {self.student_id}.")
````

### Exercise 10 — Method that changes the object

Create a `ReactionTimesAnalyzer` object from:

```python
rts = [0.41, 0.52, 0.48]
```

Then:
1. print the initial average RT
2. append a new trial `0.60`
3. print the updated list of reaction times
4. print the updated average RT


````python
rts = [0.41, 0.52, 0.48]

analyzer = ReactionTimesAnalyzer(rts)

print(analyzer.get_average_rt())
analyzer.append_new_trial(0.60)
print(analyzer.reaction_times)
print(analyzer.get_average_rt())
````

### Exercise 11 — Copy behavior inside the object

### Exercise 17 — Instance attribute vs class attribute

Create two `DemoClass` objects called `a` and `b`. (Note: `.class_attribute` and `.instance_attribute` are lists)

Then:
1. append `"shared"` to `DemoClass.class_attribute`
2. append `"local_a"` to `a.instance_attribute`
3. append `"local_b"` to `b.instance_attribute`

Print:
- `DemoClass.class_attribute`
- `a.class_attribute`
- `b.class_attribute`
- `a.instance_attribute`
- `b.instance_attribute`


````python
a = DemoClass()
b = DemoClass()

DemoClass.class_attribute.append("shared")
a.instance_attribute.append("local_a")
b.instance_attribute.append("local_b")

print(DemoClass.class_attribute)
print(a.class_attribute)
print(b.class_attribute)
print(a.instance_attribute)
print(b.instance_attribute)
````

### Exercise 23 — Understanding *args and **kwargs

Given the function below:

```python
def arbitrary_inputs_function(first_argument, *args, **kwargs):
    print(f"Captured argument: {first_argument}")
    print(f"args ({type(args)}): {args}")
    print(f"kwargs ({type(kwargs)}): {kwargs}")
```

Call it so that:
- `first_argument = "subject"`
- `args = (10, 20)`
- `kwargs = {"condition": "patient", "session": 2}`


````python
def arbitrary_inputs_function(first_argument, *args, **kwargs):
    print(f"Captured argument: {first_argument}")
    print(f"args ({type(args)}): {args}")
    print(f"kwargs ({type(kwargs)}): {kwargs}")

# Write your function call here
arbitrary_inputs_function("subject", 10, 20, condition="patient", session=2)
````
