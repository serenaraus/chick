## Exercise 6 — Choose the better error-handling strategy

You want to convert user input to an integer.

**Option A**
```python
number = int(user_input)
print(number)
```

**Option B**
```python
try:
    number = int(user_input)
    print(number)
except ValueError:
    print("That is not a valid integer.")
```

Tasks:
1. Which option is safer when `user_input` may contain invalid text?
2. Give one example input for which Option A fails.
3. Explain why Option B is preferable in that case.

````python
# ANSWERS
# 1: Option B is safer because it does not stop the whole process,
#    because "try - except" handles the error internally and the code continues wotking
# 2: Option A fails when "user_input" is not a numeric variable, such as a text (e.g. "Serena")
user_input = "Serena"
number = int(user_input)
print(number)
````

````python
# 3: Option B is preferable because it finds the error, in this case prints a message (that you can decide)
#     and still allows the rest of the code to keep running without crashing.
user_input = "Serena"
try:
    number = int(user_input)
    print(number)
except ValueError:
    print("That is not a valid integer.")
````

## Exercise 8 — Handle two different exception types

Complete the code so that it:
- catches `ZeroDivisionError` and prints `"Division by zero is not allowed."`
- catches `TypeError` and prints `"Incompatible types for division."`

Test it with:
- `a = 10, b = 0`
- `a = "10", b = 2`

````python
a = 10
b = 0

try:
    result = a / b
    print(result)
except ZeroDivisionError:
    print("Division by zero is not allowed.")
except TypeError:
    print("Incompatible types for division.")
````

````python
a = "10"
b = 2

try:
    result = a / b
    print(result)
except ZeroDivisionError:
    print("Division by zero is not allowed.")
except TypeError:
    print("Incompatible types for division.")
````

## Exercise 9 — Prevent a `KeyError`

You have a dictionary of subject groups.

Tasks:
1. Run the code and identify the exception.
2. Rewrite it using `try` / `except` so that, if the key is missing, it prints:
   `"Subject not found."`

````python
groups = {"s01": "control",
          "s02": "patient",
          "s03": "control"}

subject = "s04"

print(groups[subject])

# KeyError because "s04" does not exist in "groups"
````

````python
# New version
groups = {"s01": "control",
          "s02": "patient",
          "s03": "control"}
subject = "s04"

try:
    print(groups[subject])
except KeyError:
    print("Subject not found.")
````

## Final reflection

Answer briefly:

1. What is the difference between a `SyntaxError` and an exception? A SyntaxError is a grammar mistake that stops the code without running it. An exeption is a mistake that happened while the code was already running.
2. Why is reading a traceback from bottom to top useful? Because the bottom line tells you what is the mistake (most important information that I need first), while the lines above tell you where to find it (so in which line).
3. Why is catching a specific exception usually better than writing a generic `except`? It prevents you from accidentally hiding other mistakes you weren't expecting, making your code easier to fix.
