### Exercise 19 — Slices are usually views

Run the code below and explain why both arrays change.

Then rewrite the example using `.copy()` so that only the sliced array changes.

```python

```

````python
a = np.array([1, 2, 3, 4, 5, 6])

# First: reproduce the view behavior
b = a[1:5]
b[0] = 99
print(a)
print(b)

# Then: create an independent copy and modify it
a2 = np.array([1, 2, 3, 4, 5, 6])
b2 = a2[1:5].copy()
b2[0] = 99
print(a2)
print(b2)
````

### Exercise 23 — Mixed challenge: clean and inspect an array

Create a `(4, 5)` array of random integers from `-5` to `9`.

### Exercise 24 — Choose the better NumPy solution

You want to select all positive values from an array.

**Option A**
```python
v = np.array([-2, 3, -1, 5, 0, 4])
positive = []
for x in v:
    if x > 0:
        positive.append(x)
positive = np.array(positive)
print(positive)
```

**Option B**
```python
v = np.array([-2, 3, -1, 5, 0, 4])
positive = v[v > 0]
print(positive)
```

Tasks:
1. run both versions
2. verify that they produce the same values
3. decide which one is more idiomatic NumPy for this task
4. write one sentence explaining why

````python
# Option A
v = np.array([-2, 3, -1, 5, 0, 4])
positive_list = []
for x in v:
    if x > 0:
        positive_list.append(x)
positive_a = np.array(positive_list)

# Option B
v = np.array([-2, 3, -1, 5, 0, 4])
positive_b = v[v > 0]

print(positive_a)
print(positive_b)

# Same values?
print(np.array_equal(positive_a,positive_b))

# Add a short markdown answer below: Which solution is more idiomatic NumPy and why?
# Option B is better and more idiomatic because it uses vectorization therefore faster than passing through a for loop.
````
